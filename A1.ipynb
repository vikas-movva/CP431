```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('PrimeGapComputeTime.csv')
df
```

```python
df['time_saved'] = df['compute_time'].shift(1) - df['compute_time']
df
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='num_cores', y='time_saved', data=df)
plt.xlabel('Number of Cores')
plt.ylabel('Time Saved (seconds)')
plt.title('Time Saved per CPU Core Added')
plt.show()
```
